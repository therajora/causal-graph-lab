# Introducao

Este notebook apresenta a diferenca entre associacao e causalidade e o papel de DAGs causais como linguagem de suposicoes.

## Ideias centrais

- Um DAG causal codifica suposicoes direcionais entre variaveis.
- Confundimento surge com causas comuns de X e Y.
- Intervencao `do(X=x)` nao e a mesma coisa que observar `X=x`.

## Leitura sugerida

- Pearl, *Causality*
- Hernan & Robins, *What If*
